1. Explain the Key Features of Python That Make It a Popular Choice for Programming
Python is popular because of its versatility, readability, and strong community support. Key features include:

Easy to Read and Write: Python’s syntax is clear and readable, resembling plain English, making it ideal for beginners.

Dynamically Typed: Variables don’t need explicit type declarations, simplifying code writing and modification.

Extensive Standard Library: Python’s library includes modules for file I/O, web development, data processing, and more.

Cross-Platform Compatibility: Code written in Python runs on multiple operating systems without modifications.

Interpreted Language: Python executes code line-by-line, facilitating easier debugging and faster development.

Object-Oriented and Functional Programming: Python supports both object-oriented and functional programming paradigms.

Strong Community and Support: Python has an active community contributing to documentation, libraries, and tools.

2. Describe the Role of Predefined Keywords in Python and Provide Examples of How They Are Used in a Program
Predefined keywords in Python are reserved words that have specific meanings and cannot be used as variable or function names. These keywords control the flow, perform operations, and define structure within a program.

Examples include if, else, for, while, return, import.

Example:


# Example using 'if' and 'else' keywords
age = 18
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
In this example, if and else control the execution of code based on the condition specified.

3. Compare and Contrast Mutable and Immutable Objects in Python with Examples
In Python, objects are either mutable (can be changed) or immutable (cannot be changed).

Mutable Objects: Can be modified after creation. Examples include lists, dictionaries, and sets.


# Mutable object example
my_list = [1, 2, 3]
my_list.append(4)  # Modifies the list by adding a new element
print(my_list)  # Output: [1, 2, 3, 4]
Immutable Objects: Cannot be modified after creation. Examples include strings, tuples, and integers.


# Immutable object example
my_string = "Hello"
my_string += " World"  # Creates a new string rather than modifying the original
print(my_string)  # Output: "Hello World"
Mutable objects allow in-place modifications, while immutable objects require creating new instances to change values.

4. Explain the Concept of Type Casting in Python with Examples
Type casting is converting one data type to another. Python provides built-in functions for casting, such as:

int(): Converts to an integer
float(): Converts to a floating-point number
str(): Converts to a string
list(): Converts to a list
Examples:

# Integer to string casting
num = 10
num_str = str(num)  # Converts integer to string
print(type(num_str))  # Output: <class 'str'>

# String to integer casting
num_str = "25"
num_int = int(num_str)  # Converts string to integer
print(num_int + 5)  # Output: 30
Type casting is useful for performing operations that require specific data types.

5. Discuss the Different Types of Operators in Python and Provide Examples of How They Are Used
Python has several types of operators, each performing different operations:

Arithmetic Operators: Used for basic mathematical operations, e.g., +, -, *, /, %, ** (exponent), // (floor division).


print(5 + 3)  # Output: 8
print(5 ** 2)  # Output: 25 (exponent)
Comparison Operators: Compare two values and return a Boolean result, e.g., ==, !=, >, <, >=, <=.

print(5 > 3)  # Output: True
Logical Operators: Combine conditional statements, e.g., and, or, not.

print(True and False)  # Output: False
Assignment Operators: Assign values to variables, e.g., =, +=, -=, *=, /=.


x = 5
x += 3  # Equivalent to x = x + 3
print(x)  # Output: 8
Operators allow performing calculations, comparisons, and assignments within programs.

6. How Do Conditional Statements Work in Python? Illustrate with Examples
Conditional statements execute code based on conditions. Common conditional statements include if, elif, and else.

Example:


number = 10
if number > 0:
    print("Positive number")
elif number < 0:
    print("Negative number")
else:
    print("Zero")
In this example, Python evaluates each condition and executes the block of code that meets the specified condition.

7. Describe the Different Types of Loops in Python and Their Use Cases with Examples
Python has two primary types of loops: for and while.

for Loop: Iterates over a sequence (like a list, tuple, string, or range).

for i in range(5):
    print(i)
Output:

0
1
2
3
4
while Loop: Repeats a block of code as long as a condition is true.


count = 0
while count < 5:
    print(count)
    count += 1
Output:

0
1
2
3
4
for loops are typically used when the number of iterations is known, while while loops are better suited for situations where the loop continues until a condition is met.
